```python
x=np.arange(0,6.35,0.1)
y=np.sin(x)
plt.grid(True)
plt.xlabel("This is ")
plt.plot(x,y)
```

```python
z=np.arange(0,6.35,0.1)
y=np.sin(z)
x=np.cos(z)
plt.grid(True)
plt.title("3 Phase Waveform")
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.plot(y,label="Phase-A")
plt.plot(x,label="Phase-B")
plt.plot(-x,label="Phase-C")
plt.legend(loc="upper left")
plt.savefig('3phase.pdf')
plt.show()
```
