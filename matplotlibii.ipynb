# Control colors

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.arange(1, 3)
plt.plot(y, 'y');
plt.plot(y+1, 'm');
plt.plot(y+2, 'c');
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.arange(1, 3)
plt.plot(y, 'y', y+1, 'm', y+2, 'c');
plt.show()
```

# Control line styles

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.arange(1, 3)
plt.plot(y, '--', y+1, '-.', y+2, ':');
plt.show()
```

# Control marker styles

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.arange(1, 3, 0.2)
plt.plot(y, 'x', y+0.5, 'o', y+1, 'D', y+1.5, '^', y+2, 's');
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.arange(1, 3, 0.3)
plt.plot(y, 'cx--', y+1, 'mo:', y+2, 'kp-.');
plt.show()
```

# Finer control with keyword arguments

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.arange(1, 3, 0.3)
plt.plot(y, color='blue', linestyle='dashdot', linewidth=4, marker='o', markerfacecolor='red', markeredgecolor='black',
markeredgewidth=3, markersize=12)
plt.show()
```

# Handling X and Y ticks

```python
import matplotlib.pyplot as plt
x = [5, 3, 7, 2, 4, 1]
plt.plot(x);
plt.xticks(range(len(x)), ['a', 'b', 'c', 'd', 'e', 'f'])
plt.yticks(range(1, 8, 2))
plt.show()
```

# Histogram charts

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.random.randn(1000)
plt.hist(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.random.randn(1000)
plt.hist(y , 25)
plt.show()
```

# Bar charts

```python
import matplotlib.pyplot as plt
plt.bar([1, 2, 3], [3, 2, 5]);
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
dict = {'A': 40, 'B': 70, 'C': 30, 'D': 85}
for i, key in enumerate(dict): plt.bar(i, dict[key])
plt.xticks(np.arange(5), dict.keys())
plt.yticks(dict.values())
plt.show()
```
